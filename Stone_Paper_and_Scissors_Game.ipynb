```python
# Rock, Paper and scissor game 

rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''

#Write your code below this line ðŸ‘‡
import random
computer=random.randint(0,2)

user_inp=int(input("What do you choose? Type 0 for Rock, 1 for Paper and 2 for scissor\n"))

if user_inp==0:
  print(rock)
  #program for computer round
  print("Computer choose:")
  if computer==0:
   print(rock)
  elif computer==1:
   print(paper)
  elif computer==2:
   print(scissors)
  
elif user_inp==1:
  print(paper)
  #program for computer round
  print("Computer choose:")
  if computer==0:
    print(rock)
  elif computer==1:
    print(paper)
  elif computer==2:
    print(scissors)

elif user_inp==2:
  print(scissors)
  #program for computer round
  print("Computer choose:")
  if computer==0:
   print(rock)
  elif computer==1:
   print(paper)
  elif computer==2:
   print(scissors)

else:
  print("Error \nPlease enter correct number")

#Rule codes

if user_inp==0 and computer==0:
  print("Match draw")
elif user_inp==0 and computer==1:
  print("You loose")
elif user_inp==0 and computer==2:
  print("You win")

elif user_inp==1 and computer==0:
  print("You win")
elif user_inp==1 and computer==1:
  print("Match draw")
elif user_inp==1 and computer==2:
  print("You loose")

elif user_inp==2 and computer==0:
  print("You loose")
elif user_inp==2 and computer==1:
  print("You win")
elif user_inp==2 and computer==2:
  print("Match draw")




```
